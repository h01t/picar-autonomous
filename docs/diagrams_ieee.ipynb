```python
#!/usr/bin/env python3
"""
DrivingCNN Architecture (PilotNet-style)
IEEE-Compliant Formatting (600 DPI, Times New Roman, Color)
"""

import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, Rectangle
import matplotlib.font_manager as fm

fm._load_fontmanager(try_read_cache=False)

# IEEE-compliant settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Liberation Serif', 'DejaVu Serif']
plt.rcParams['font.size'] = 8
plt.rcParams['axes.linewidth'] = 1.0
plt.rcParams['lines.linewidth'] = 1.0

fig_width = 7.16
fig_height = 4.5

fig, ax = plt.subplots(figsize=(fig_width, fig_height))
ax.set_xlim(0, 18)
ax.set_ylim(0, 10)
ax.axis('off')

# Actual layer definitions from model.py
conv_layers = [
    ('Input\n224×224×3', 2.2, '#e3f2fd', 'RGB Image'),
    ('Conv1\n5×5, s=2\n→24 filters', 2.0, '#bbdefb', '110×110×24'),
    ('Conv2\n5×5, s=2\n→36 filters', 1.8, '#90caf9', '53×53×36'),
    ('Conv3\n5×5, s=2\n→48 filters', 1.6, '#64b5f6', '25×25×48'),
    ('Conv4\n3×3, s=1\n→64 filters', 1.4, '#42a5f5', '23×23×64'),
    ('Conv5\n3×3, s=1\n→64 filters', 1.2, '#2196f3', '21×21×64'),
]

fc_layers = [
    ('Flatten\n28224', 1.0, '#e1bee7', 'Feature Vector'),
    ('FC1\n→100\n+Dropout', 0.8, '#ce93d8', 'ReLU + 50% Drop'),
    ('FC2\n→50\n+Dropout', 0.6, '#ba68c8', 'ReLU + 50% Drop'),
    ('Output\n→2', 0.4, '#ab47bc', '[steer, throttle]'),
]

y_center = 5
x_start = 0.3

# Draw convolutional layers
for i, (name, height_scale, color, desc) in enumerate(conv_layers):
    x = x_start + i * 1.8
    h = 3 * height_scale
    y = y_center - h/2
    
    rect = FancyBboxPatch((x, y), 1.4, h, boxstyle="round,pad=0.02",
                           facecolor=color, edgecolor='#333', linewidth=1.0)
    ax.add_patch(rect)
    ax.text(x + 0.7, y_center, name, fontsize=6, ha='center', va='center', 
            fontweight='bold', family='serif')
    
    if i < len(conv_layers) - 1:
        ax.annotate('', xy=(x + 1.6, y_center), xytext=(x + 1.4, y_center),
                   arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
    
    ax.text(x + 0.7, y - 0.3, desc, fontsize=5, ha='center', color='#666', 
            style='italic', family='serif')

# Arrow from conv to FC
ax.annotate('', xy=(x_start + len(conv_layers) * 1.8 + 0.2, y_center), 
           xytext=(x_start + len(conv_layers) * 1.8 - 0.2, y_center),
           arrowprops=dict(arrowstyle='->', color='#9C27B0', lw=1.5))

# Draw FC layers
fc_x_start = x_start + len(conv_layers) * 1.8 + 0.4
for i, (name, height_scale, color, desc) in enumerate(fc_layers):
    x = fc_x_start + i * 1.6
    h = 2.5 * height_scale
    y = y_center - h/2
    
    rect = FancyBboxPatch((x, y), 1.3, h, boxstyle="round,pad=0.02",
                           facecolor=color, edgecolor='#333', linewidth=1.0)
    ax.add_patch(rect)
    ax.text(x + 0.65, y_center, name, fontsize=6, ha='center', va='center', 
            fontweight='bold', family='serif')
    
    if i < len(fc_layers) - 1:
        ax.annotate('', xy=(x + 1.5, y_center), xytext=(x + 1.3, y_center),
                   arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
    
    ax.text(x + 0.65, y - 0.3, desc, fontsize=5, ha='center', color='#666', 
            style='italic', family='serif')

# Labels for sections
ax.text(5.5, 8.25, 'Convolutional Feature Extraction', fontsize=9, fontweight='bold', 
        ha='center', color='#1976D2', family='serif') 
ax.add_patch(Rectangle((0.2, 1.5), 10.6, 7, fill=False, edgecolor='#1976D2', 
                        linestyle='--', linewidth=1.0, alpha=0.5))

ax.text(14.5, 7.5, 'Fully Connected', fontsize=9, fontweight='bold', 
        ha='center', color='#7B1FA2', family='serif')
ax.add_patch(Rectangle((11, 2.5), 6.75, 5.5, fill=False, edgecolor='#7B1FA2', 
                        linestyle='--', linewidth=1.0, alpha=0.5))

# Preprocessing note
preprocess_box = FancyBboxPatch((0.2, 8.8), 5, 0.9, boxstyle="round,pad=0.03",
                                 facecolor='#fff3e0', edgecolor='#F39C12', linewidth=1.0)
ax.add_patch(preprocess_box)
ax.text(2.7, 9.25, 'Preprocessing: RGB → /255.0 → Transpose(C,H,W)', fontsize=7, 
        ha='center', family='serif')

# Output note
output_box = FancyBboxPatch((11, 8.8), 5, 0.9, boxstyle="round,pad=0.03",
                             facecolor='#e8f5e9', edgecolor='#27AE60', linewidth=1.0)
ax.add_patch(output_box)
ax.text(13.3, 9.25, 'Output: [steering, throttle] (unbounded)', fontsize=7, 
        ha='center', family='serif')

# Architecture summary box
summary_box = FancyBboxPatch((0.5, 0.1), 15.6, 1, boxstyle="round,pad=0.03",
                              facecolor='#fafafa', edgecolor='#666', linewidth=1.0)
ax.add_patch(summary_box)
ax.text(8, 0.75, 'Architecture Summary (inspired by NVIDIA PilotNet)', fontsize=8, 
        fontweight='bold', ha='center', family='serif')
ax.text(8, 0.25, '5 Conv layers (24→36→48→64→64) | 3 FC layers (28224→100→50→2) | '
        'Dropout 50% | ReLU activations | ~850K parameters', fontsize=7, ha='center', family='serif')

plt.tight_layout(pad=0.1)

plt.savefig('bleja/02_cnn_architecture.png', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='png')
plt.savefig('bleja/02_cnn_architecture.tiff', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='tiff')

print('Saved: 02_cnn_architecture.png and .tiff (600 DPI)')
plt.show()
```

```python
#!/usr/bin/env python3
"""
Obstacle Avoidance Algorithm Flowchart
IEEE-Compliant Formatting (600 DPI, Times New Roman, Color)
"""

import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, Polygon, Ellipse
import matplotlib.font_manager as fm

fm._load_fontmanager(try_read_cache=False)

# IEEE-compliant settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Liberation Serif', 'DejaVu Serif']
plt.rcParams['font.size'] = 8
plt.rcParams['axes.linewidth'] = 1.0
plt.rcParams['lines.linewidth'] = 1.0

fig_width = 5.0
fig_height = 7.0

fig, ax = plt.subplots(figsize=(fig_width, fig_height))
ax.set_xlim(0, 10)
ax.set_ylim(0, 14)
ax.axis('off')

def add_rect(x, y, w, h, text, color='#e3f2fd'):
    box = FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.02",
                         facecolor=color, edgecolor='#333', linewidth=1.0)
    ax.add_patch(box)
    ax.text(x + w/2, y + h/2, text, fontsize=7, ha='center', va='center', family='serif')

def add_diamond(x, y, w, h, text):
    verts = [(x, y + h/2), (x + w/2, y + h), (x + w, y + h/2), (x + w/2, y)]
    diamond = Polygon(verts, facecolor='#fff3e0', edgecolor='#333', linewidth=1.0)
    ax.add_patch(diamond)
    ax.text(x + w/2, y + h/2, text, fontsize=7, ha='center', va='center', family='serif')

def add_oval(x, y, w, h, text):
    ellipse = Ellipse((x + w/2, y + h/2), w, h, facecolor='#e8f5e9', edgecolor='#333', linewidth=1.0)
    ax.add_patch(ellipse)
    ax.text(x + w/2, y + h/2, text, fontsize=8, ha='center', va='center', family='serif')

# Flowchart elements
add_oval(3.5, 13, 3, 0.7, 'Start')
add_rect(3, 11.5, 4, 0.8, 'Read Ultrasonic Sensors\n(Front, Left, Right, Rear)', '#bbdefb')
add_diamond(3, 9.5, 4, 1.2, 'Front\n< 25cm?')
add_rect(0.5, 8, 3, 0.8, 'Check Side\nSensors', '#bbdefb')
add_diamond(0.5, 6, 3, 1.2, 'Turning\ninto\nobstacle?')
add_rect(6.5, 8, 3, 0.8, 'STOP\nMotors', '#ffcdd2')
add_diamond(6.5, 6, 3, 1.2, 'Rear\n> 30cm?')
add_rect(5.5, 4.2, 2, 0.8, 'Reverse\n(0.8s)', '#fff9c4')
add_rect(5.5, 2.8, 2, 0.8, 'Compare\nL vs R', '#bbdefb')
add_rect(5.5, 1.2, 2, 1.0, 'Turn toward\nclearer side\n(0.9s)', '#c8e6c9')
add_rect(8.2, 4.2, 1.5, 1.0, 'Turn in\nplace\n(1.2s)', '#c8e6c9')
add_rect(0.5, 4.2, 3, 0.8, 'Continue Normal\nOperation', '#c8e6c9')
add_oval(3.5, 0, 3, 0.7, 'Resume')

# Arrows
arrows = [
    ((5, 13), (5, 12.3)),
    ((5, 11.5), (5, 10.7)),
]
for start, end in arrows:
    ax.annotate('', xy=end, xytext=start, arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))

# Decision arrows with labels
ax.annotate('', xy=(2, 8.8), xytext=(3, 10.1), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.text(2.2, 9.5, 'No', fontsize=7, family='serif')

ax.annotate('', xy=(8, 8.8), xytext=(7, 10.1), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.text(7.5, 9.5, 'Yes', fontsize=7, family='serif')

ax.annotate('', xy=(2, 7.2), xytext=(2, 8), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))

ax.annotate('', xy=(2, 5), xytext=(0.5, 6.6), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.text(0.8, 5.5, 'No', fontsize=7, family='serif')

ax.annotate('', xy=(7, 8), xytext=(3.5, 6.6), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.text(4.2, 7, 'Yes', fontsize=7, family='serif')

ax.annotate('', xy=(8, 7.2), xytext=(8, 8), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))

ax.annotate('', xy=(6.5, 5), xytext=(6.5, 6.6), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.text(6.8, 5.5, 'Yes', fontsize=7, family='serif')

ax.annotate('', xy=(8.95, 5.2), xytext=(9.5, 6.6), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.text(9.3, 5.8, 'No', fontsize=7, family='serif')

ax.annotate('', xy=(6.5, 3.6), xytext=(6.5, 4.2), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.annotate('', xy=(6.5, 2.2), xytext=(6.5, 2.8), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.annotate('', xy=(5, 0.7), xytext=(6.5, 1.2), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.annotate('', xy=(7.5, 3.15), xytext=(9, 4.2), arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))

plt.tight_layout(pad=0.1)

plt.savefig('bleja/05_obstacle_avoidance_flowchart.png', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='png')
plt.savefig('bleja/05_obstacle_avoidance_flowchart.tiff', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='tiff')

print('Saved: 05_obstacle_avoidance_flowchart.png and .tiff (600 DPI)')
plt.show()
```

```python
#!/usr/bin/env python3
"""
Training Data Distribution Analysis
IEEE-Compliant Formatting (600 DPI, Times New Roman, Color)
NOTE: Update the CSV path to your actual labels.csv location
"""

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm

fm._load_fontmanager(try_read_cache=False)

# IEEE-compliant settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Liberation Serif', 'DejaVu Serif']
plt.rcParams['font.size'] = 8
plt.rcParams['axes.linewidth'] = 1.0
plt.rcParams['lines.linewidth'] = 1.0

# Load data - UPDATE THIS PATH to your actual labels.csv location
try:
    df = pd.read_csv('/Users/grmim/Dev/py/final/dataset/labels.csv')
except FileNotFoundError:
    # Create sample data if file not found
    np.random.seed(42)
    n_samples = 20000
    df = pd.DataFrame({
        'steering': np.random.normal(0, 0.3, n_samples),
        'throttle': np.random.uniform(0.2, 0.8, n_samples)
    })
    print("Note: Using sample data. Update path to your labels.csv for actual data.")

fig_width = 7.16
fig_height = 5.5

fig = plt.figure(figsize=(fig_width, fig_height))
gs = gridspec.GridSpec(2, 2, hspace=0.35, wspace=0.35)

# 1. Steering Distribution Histogram
ax1 = fig.add_subplot(gs[0, 0])
ax1.hist(df['steering'], bins=50, color='#3498DB', edgecolor='white', alpha=0.8)
ax1.axvline(x=0, color='#E74C3C', linestyle='--', label='Neutral', linewidth=1.0)
ax1.set_xlabel('Steering Value', fontsize=8, family='serif')
ax1.set_ylabel('Frequency', fontsize=8, family='serif')
ax1.set_title('Steering Distribution (n={:,})'.format(len(df)), fontsize=9, 
              fontweight='bold', family='serif')
ax1.legend(fontsize=7)
ax1.tick_params(labelsize=7)

stats_text = f'μ = {df["steering"].mean():.3f}\nσ = {df["steering"].std():.3f}'
ax1.text(0.95, 0.95, stats_text, transform=ax1.transAxes, ha='right', va='top',
         fontsize=7, family='serif', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 2. Throttle Distribution Histogram
ax2 = fig.add_subplot(gs[0, 1])
ax2.hist(df['throttle'], bins=50, color='#27AE60', edgecolor='white', alpha=0.8)
ax2.set_xlabel('Throttle Value', fontsize=8, family='serif')
ax2.set_ylabel('Frequency', fontsize=8, family='serif')
ax2.set_title('Throttle Distribution', fontsize=9, fontweight='bold', family='serif')
ax2.tick_params(labelsize=7)

stats_text = f'μ = {df["throttle"].mean():.3f}\nσ = {df["throttle"].std():.3f}'
ax2.text(0.05, 0.95, stats_text, transform=ax2.transAxes, ha='left', va='top',
         fontsize=7, family='serif', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 3. Steering Category Pie Chart
ax3 = fig.add_subplot(gs[1, 0])
left = len(df[df['steering'] < -0.1])
straight = len(df[(df['steering'] >= -0.1) & (df['steering'] <= 0.1)])
right = len(df[df['steering'] > 0.1])

sizes = [left, straight, right]
labels = [f'Left Turn\n(< -0.1)\n{left:,}', 
          f'Straight\n(-0.1 to 0.1)\n{straight:,}', 
          f'Right Turn\n(> 0.1)\n{right:,}']
colors_pie = ['#3498DB', '#ECF0F1', '#F39C12']
explode = (0.02, 0.02, 0.02)

ax3.pie(sizes, explode=explode, labels=labels, colors=colors_pie, autopct='%1.1f%%',
        shadow=False, startangle=90, textprops={'fontsize': 7, 'family': 'serif'})
ax3.set_title('Steering Command Categories', fontsize=9, fontweight='bold', family='serif')

# 4. 2D Density Plot (Steering vs Throttle)
ax4 = fig.add_subplot(gs[1, 1])
h = ax4.hist2d(df['steering'], df['throttle'], bins=40, cmap='YlOrRd')
cbar = plt.colorbar(h[3], ax=ax4)
cbar.set_label('Count', fontsize=7, family='serif')
cbar.ax.tick_params(labelsize=6)
ax4.set_xlabel('Steering', fontsize=8, family='serif')
ax4.set_ylabel('Throttle', fontsize=8, family='serif')
ax4.set_title('Steering vs Throttle Density', fontsize=9, fontweight='bold', family='serif')
ax4.tick_params(labelsize=7)

plt.tight_layout(pad=0.5)

plt.savefig('bleja/06_training_data_distribution.png', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='png')
plt.savefig('bleja/06_training_data_distribution.tiff', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='tiff')

print('Saved: 06_training_data_distribution.png and .tiff (600 DPI)')
plt.show()
```

```python
#!/usr/bin/env python3
"""
Communication Sequence Diagram
IEEE-Compliant Formatting (600 DPI, Times New Roman, Color)
"""

import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
import matplotlib.font_manager as fm

fm._load_fontmanager(try_read_cache=False)

# IEEE-compliant settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Liberation Serif', 'DejaVu Serif']
plt.rcParams['font.size'] = 8
plt.rcParams['axes.linewidth'] = 1.0
plt.rcParams['lines.linewidth'] = 1.0

COLORS = {
    'flutter': '#02569B',
    'success': '#27AE60',
    'box_ctrl': '#e3f2fd',
    'box_tele': '#e8f5e9',
    'line': '#000000',
}

fig_width = 7.16
fig_height = 5.5

fig, ax = plt.subplots(figsize=(fig_width, fig_height))
ax.set_xlim(0, 14)
ax.set_ylim(0, 10)
ax.axis('off')

# Lifelines with clearer labels
entities = [
    (2, 'Flutter App'),
    (6, 'Control Socket\n(Port 5005)'),
    (10, 'Telemetry Socket\n(Port 5006)'),
]

for x, name in entities:
    box = FancyBboxPatch((x-1, 9.0), 2, 0.8, boxstyle="round,pad=0.02",
                         facecolor='#f0f0f0', edgecolor='#000000', linewidth=1.0)
    ax.add_patch(box)
    ax.text(x, 9.4, name, ha='center', va='center', fontsize=8, fontweight='bold',
            family='serif')
    ax.plot([x, x], [0.8, 9.0], 'k--', alpha=0.6, linewidth=0.8)

# Connection phase header
y = 8.3
ax.text(6, y + 0.3, 'Connection Phase', fontsize=9, ha='center', 
        fontweight='bold', family='serif', style='italic')

# TCP Connect arrows
y = 7.8
ax.annotate('', xy=(6-0.15, y), xytext=(2+0.15, y),
           arrowprops=dict(arrowstyle='->', color=COLORS['line'], lw=1.2))
ax.text(4, y+0.15, 'TCP Connect', fontsize=8, ha='center', family='serif')

y -= 0.5
ax.annotate('', xy=(10-0.15, y), xytext=(2+0.15, y),
           arrowprops=dict(arrowstyle='->', color=COLORS['line'], lw=1.2))
ax.text(6, y+0.15, 'TCP Connect', fontsize=8, ha='center', family='serif')

# Loop indicator
y -= 0.7
loop_top = y + 0.3
loop_bottom = 1.2
loop_box = FancyBboxPatch((1.3, loop_bottom), 9.4, loop_top - loop_bottom, 
                           boxstyle="round,pad=0.02",
                           facecolor='none', edgecolor='#666666', 
                           linewidth=1.0, linestyle='--')
ax.add_patch(loop_box)
ax.text(1.5, loop_top - 0.15, 'loop', fontsize=7, family='serif',
        style='italic', color='#444444', fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.1', facecolor='white', edgecolor='none'))
ax.text(6, loop_top - 0.15, '[50ms interval]', fontsize=7, ha='center',
        family='serif', style='italic', color='#444444')

# Control command (first)
y -= 0.6
ax.annotate('', xy=(6-0.15, y), xytext=(2+0.15, y),
           arrowprops=dict(arrowstyle='->', color=COLORS['flutter'], lw=1.5))
ax.text(4, y+0.15, 'JSON: {steering, throttle, mode}', fontsize=7, 
        ha='center', family='serif', color=COLORS['flutter'])

# Telemetry response (first)
y -= 0.6
ax.annotate('', xy=(2+0.15, y), xytext=(10-0.15, y),
           arrowprops=dict(arrowstyle='->', color=COLORS['success'], lw=1.5))
ax.text(6, y+0.15, 'JSON: {fps, steering, throttle, sensors, obstacle}', 
        fontsize=7, ha='center', family='serif', color=COLORS['success'])

# Additional loop iterations
for i in range(3):
    y -= 0.55
    ax.annotate('', xy=(6-0.15, y), xytext=(2+0.15, y),
               arrowprops=dict(arrowstyle='->', color=COLORS['flutter'], lw=1.2))
    y -= 0.45
    ax.annotate('', xy=(2+0.15, y), xytext=(10-0.15, y),
               arrowprops=dict(arrowstyle='->', color=COLORS['success'], lw=1.2))

# Continuation indicator
y -= 0.4
ax.text(6, y, '...', fontsize=12, ha='center', color='#666666')

# Legend
legend_x = 11.5
legend_y = 8.5
ax.plot([legend_x, legend_x + 0.6], [legend_y, legend_y], 
        color=COLORS['flutter'], lw=2)
ax.text(legend_x + 0.75, legend_y, 'Control', fontsize=7, va='center', family='serif')
ax.plot([legend_x, legend_x + 0.6], [legend_y - 0.4, legend_y - 0.4], 
        color=COLORS['success'], lw=2)
ax.text(legend_x + 0.75, legend_y - 0.4, 'Telemetry', fontsize=7, va='center', family='serif')

# Data format boxes
ctrl_box = FancyBboxPatch((11, 5.2), 2.8, 2.2, boxstyle="round,pad=0.03",
                           facecolor=COLORS['box_ctrl'], edgecolor='#000000', linewidth=1.0)
ax.add_patch(ctrl_box)
ax.text(12.4, 7.15, 'Control Packet', fontsize=8, ha='center', fontweight='bold', family='serif')
ctrl_json = '{\n "steering": 0.15,\n "throttle": 0.5,\n "mode": "manual"\n}'
ax.text(12.4, 6.0, ctrl_json, fontsize=6, ha='center', va='center',
        family='monospace', linespacing=1.2)

tele_box = FancyBboxPatch((11, 2.0), 2.8, 2.8, boxstyle="round,pad=0.03",
                           facecolor=COLORS['box_tele'], edgecolor='#000000', linewidth=1.0)
ax.add_patch(tele_box)
ax.text(12.4, 4.55, 'Telemetry Packet', fontsize=8, ha='center', fontweight='bold', family='serif')
tele_json = '{\n "fps": 20,\n "steering": 0.0,\n "sensors": {...},\n "obstacle": false\n}'
ax.text(12.4, 3.2, tele_json, fontsize=6, ha='center', va='center',
        family='monospace', linespacing=1.2)

plt.tight_layout(pad=0.1)

plt.savefig('bleja/07_communication_sequence.png', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='png')
plt.savefig('bleja/07_communication_sequence.tiff', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='tiff')

print('Saved: 07_communication_sequence.png and .tiff (600 DPI)')
plt.show()
```

```python
#!/usr/bin/env python3
"""
Training Pipeline with Data Augmentation Diagram
IEEE-Compliant Formatting (600 DPI, Times New Roman, Color)
"""

import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
import matplotlib.font_manager as fm

fm._load_fontmanager(try_read_cache=False)

# IEEE-compliant settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'Liberation Serif', 'DejaVu Serif']
plt.rcParams['font.size'] = 8
plt.rcParams['axes.linewidth'] = 1.0
plt.rcParams['lines.linewidth'] = 1.0

fig_width = 7.16
fig_height = 5.5

fig, ax = plt.subplots(figsize=(fig_width, fig_height))
ax.set_xlim(0, 16)
ax.set_ylim(0, 12)
ax.axis('off')

# === DATA LOADING SECTION ===
ax.text(2.5, 11.25, 'Data Loading & Augmentation', fontsize=9, fontweight='bold', 
        ha='center', color='#1976D2', family='serif')

data_steps = [
    (0.5, 9.5, 'Raw Dataset\n~20,000 frames\n(labels.csv)', '#e3f2fd'),
    (3, 9.5, 'Train/Val Split\n80% / 20%\n(random_state=42)', '#bbdefb'),
]

for x, y, text, color in data_steps:
    box = FancyBboxPatch((x, y), 2.2, 1.5, boxstyle="round,pad=0.03",
                         facecolor=color, edgecolor='#333', linewidth=1.0)
    ax.add_patch(box)
    ax.text(x + 1.1, y + 0.75, text, fontsize=7, ha='center', va='center', family='serif')

ax.annotate('', xy=(3, 10.25), xytext=(2.7, 10.25),
           arrowprops=dict(arrowstyle='->', color='#333', lw=1.2))

# === AUGMENTATION SECTION ===
aug_box = FancyBboxPatch((5.8, 8.8), 3.5, 2.8, boxstyle="round,pad=0.03",
                          facecolor='#fff3e0', edgecolor='#F39C12', linewidth=1.5)
ax.add_patch(aug_box)
ax.text(7.55, 11.3, 'Data Augmentation', fontsize=8, fontweight='bold', ha='center', 
        color='#F39C12', family='serif')
ax.text(7.55, 11, '(Training Only)', fontsize=7, ha='center', style='italic', family='serif')

augmentations = [
    'RandomBrightnessContrast (±20%)',
    'HueSaturationValue',
    'GaussianNoise (p=0.2)',
    'HorizontalFlip (p=0.5)*',
]
for i, aug in enumerate(augmentations):
    ax.text(6.1, 10.5 - i*0.4, f'• {aug}', fontsize=6, family='serif')

ax.text(6.1, 7.95, '*Flips also invert steering', fontsize=5, style='italic', 
        color='#666', family='serif')

ax.annotate('', xy=(5.8, 10.25), xytext=(5.2, 10.25),
           arrowprops=dict(arrowstyle='->', color='#333', lw=1.2))

# === MODEL SECTION ===
ax.text(11.5, 11.25, 'Model & Training', fontsize=9, fontweight='bold', 
        ha='center', color='#7B1FA2', family='serif')

model_box = FancyBboxPatch((10, 9.5), 3, 1.5, boxstyle="round,pad=0.03",
                            facecolor='#f3e5f5', edgecolor='#7B1FA2', linewidth=1.0)
ax.add_patch(model_box)
ax.text(11.5, 10.6, 'DrivingCNN', fontsize=8, fontweight='bold', ha='center', family='serif')
ax.text(11.5, 10.1, '5 Conv + 3 FC layers\nDropout 50%', fontsize=6, ha='center', family='serif')

ax.annotate('', xy=(10, 10.25), xytext=(9.3, 10.25),
           arrowprops=dict(arrowstyle='->', color='#333', lw=1.2))

# === TRAINING LOOP ===
loop_box = FancyBboxPatch((1, 4), 12, 4, boxstyle="round,pad=0.05",
                           facecolor='#fafafa', edgecolor='#27AE60', linewidth=1.5)
ax.add_patch(loop_box)
ax.text(7, 7.7, 'Training Loop (max 100 epochs)', fontsize=9, fontweight='bold', 
        ha='center', color='#27AE60', family='serif')

# Training components
train_components = [
    (1.5, 5.8, 'DataLoader\nbatch_size=64\nshuffle=True', '#e8f5e9'),
    (4.2, 5.8, 'Forward Pass\npred = model(x)', '#c8e6c9'),
    (6.9, 5.8, 'MSE Loss\nloss(pred, y)', '#a5d6a7'),
    (9.6, 5.8, 'Backward +\nAdam Update\nlr=1e-4', '#81c784'),
]

for x, y, text, color in train_components:
    box = FancyBboxPatch((x, y), 2.2, 1.5, boxstyle="round,pad=0.02",
                         facecolor=color, edgecolor='#333', linewidth=1.0)
    ax.add_patch(box)
    ax.text(x + 1.1, y + 0.75, text, fontsize=6, ha='center', va='center', family='serif')

for i in range(3):
    ax.annotate('', xy=(4.2 + i*2.7, 6.5), xytext=(3.7 + i*2.7, 6.5),
               arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))

# Validation & Callbacks
val_box = FancyBboxPatch((1.5, 4.3), 4, 1.2, boxstyle="round,pad=0.02",
                          facecolor='#fff9c4', edgecolor='#F39C12', linewidth=1.0)
ax.add_patch(val_box)
ax.text(3.5, 5.2, 'Validation', fontsize=8, fontweight='bold', ha='center', family='serif')
ax.text(3.5, 4.7, 'No augmentation | Compute val_loss', fontsize=6, ha='center', family='serif')

lr_box = FancyBboxPatch((6, 4.3), 3.2, 1.2, boxstyle="round,pad=0.02",
                         facecolor='#e1bee7', edgecolor='#9C27B0', linewidth=1.0)
ax.add_patch(lr_box)
ax.text(7.6, 5.2, 'LR Scheduler', fontsize=8, fontweight='bold', ha='center', family='serif')
ax.text(7.6, 4.7, 'ReduceOnPlateau\nfactor=0.5, patience=3', fontsize=6, ha='center', family='serif')

early_box = FancyBboxPatch((9.7, 4.3), 3, 1.2, boxstyle="round,pad=0.02",
                            facecolor='#ffcdd2', edgecolor='#E74C3C', linewidth=1.0)
ax.add_patch(early_box)
ax.text(11.2, 5.2, 'Early Stopping', fontsize=8, fontweight='bold', ha='center', family='serif')
ax.text(11.2, 4.7, 'patience=7 epochs\nSave best model', fontsize=6, ha='center', family='serif')

# === OUTPUT ===
output_steps = [
    (3, 1.8, 'model_best.pth\n(lowest val_loss)', '#c8e6c9'),
    (6, 1.8, 'Dynamic\nQuantization', '#e1bee7'),
    (9, 1.8, 'model.pt\n(TorchScript)', '#b2dfdb'),
]

for x, y, text, color in output_steps:
    box = FancyBboxPatch((x, y), 2.5, 1.2, boxstyle="round,pad=0.02",
                         facecolor=color, edgecolor='#333', linewidth=1.0)
    ax.add_patch(box)
    ax.text(x + 1.25, y + 0.6, text, fontsize=7, ha='center', va='center', family='serif')

ax.annotate('', xy=(4, 3), xytext=(4, 3.95),
           arrowprops=dict(arrowstyle='->', color='#27AE60', lw=1.2))
ax.annotate('', xy=(6, 2.4), xytext=(5.5, 2.4),
           arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))
ax.annotate('', xy=(9, 2.4), xytext=(8.5, 2.4),
           arrowprops=dict(arrowstyle='->', color='#333', lw=1.0))

ax.text(13, 2.5, 'Deploy to\nRaspberry Pi', fontsize=7, ha='center', style='italic', family='serif')
ax.annotate('', xy=(14, 2.4), xytext=(11.5, 2.4),
           arrowprops=dict(arrowstyle='->', color='#009688', lw=1.5))

plt.tight_layout(pad=0.1)

plt.savefig('bleja/16_training_pipeline_augmentation.png', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='png')
plt.savefig('bleja/16_training_pipeline_augmentation.tiff', dpi=600, bbox_inches='tight',
            pad_inches=0.02, facecolor='white', edgecolor='none', format='tiff')

print('Saved: 16_training_pipeline_augmentation.png and .tiff (600 DPI)')
plt.show()
```
